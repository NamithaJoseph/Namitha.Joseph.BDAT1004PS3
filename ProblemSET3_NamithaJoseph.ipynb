#Step 2

```python
data=pd.read_csv('D:\\wind.txt',sep="\s+",parse_dates=[[0,1,2]])

data.head()
```

#Step 3

```python
def fixYears(var):
    if var.year > 1979:
        year=var.year-100
    else:
        year = var.year
    return datetime.date(year,var.month,var.day)

```

```python
data['Yr_Mo_Dy']=data['Yr_Mo_Dy'].apply(fixYears)
data.head()
```

# Question 6

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
marriages=pd.read_csv('D:\\us-marriages-divorces-1867-2014.csv')
marriages.head()
```

```python
years=marriages.Year.values
marraiges=marriages.Marriages_per_1000.values
divorces=marriages.Divorces_per_1000.values

plt.plot(years,marraiges,color="Red")
plt.plot(years,divorces,color="Blue")
plt.xlabel("Count of Years")
plt.ylabel("Marriages and divoces per capita")
plt.title("Number of Marriages and Divorces per capita in the U.S. between 1867 and 2014. \n\n\n")
plt.show()
```

# Question 7 

```python
years_1900=marriages.Year>=1900
years_2000=marriages.Year<=2000
Data_1900_2000=marriages[years_1900 & years_2000]
plt.bar(Data_1900_2000['Year'],Data_1900_2000['Marriages_per_1000'],color="Orange")
plt.bar(Data_1900_2000['Year'],Data_1900_2000['Divorces_per_1000'],color="Blue")
plt.title("Number of Marraiges and Divorces per capita in US between 1900 to 2000 \n\n\n")
plt.xlabel("Between 1900 and 2000")
plt.ylabel("Marriages and Divorces per capita in US")
plt.show()
```
